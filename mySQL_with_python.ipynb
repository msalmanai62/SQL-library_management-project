```python
import pandas as pd
import mysql.connector as conn
```

## Setting up and checking connection

```python
connection = conn.connect(
    host="127.0.0.1",
    user="root",
    password="admin",
    database="library_management"
)
```

```python
# we have written a function to execute queries and show results
def exec__show(query, show_res=False):
    try:
        cursor = connection.cursor()
        # Execute the query
        cursor.execute(query)
        
        # Fetch all results to avoid unread result errors
        results = cursor.fetchall()
        # Optionally display the results
        if show_res:
            return pd.DataFrame(results, columns=[desc[0] for desc in cursor.description])
        cursor.close()
    except conn.Error as err:
        print(f"Error: {err}")
    finally:
        pass
```

```python
exec__show("select * from books limit 10", show_res=True)
```

```python
exec__show("create table temp2 (id int primary key, age int(2))", show_res=False)
```

```python
## Task 13: Identify Members with Overdue Books
## Write a query to identify members who have overdue books (assume a 30-day return period). Display the member's_id, member's name, book  title, issue date, and days overdue.


exec__show("select mem.member_id as mem_id, mem.member_name, iss.issued_date as is_date, bk.book_title, bk.isbn, rs.return_date as ret_date, DATEDIFF(COALESCE(rs.return_date, CURDATE()), iss.issued_date) AS no_of_days from members as mem join issued_status as iss on mem.member_id=iss.issued_member_id join books as bk on iss.issued_book_isbn=bk.isbn left join return_status as rs on iss.issued_id=rs.issued_id where rs.return_date is null and DATEDIFF(COALESCE(rs.return_date, CURDATE()), iss.issued_date) > 50", show_res=True)
```

```python
exec__show("""
SELECT
       mem.member_id AS mem_id,
       mem.member_name,
       iss.issued_date AS is_date,
       bk.book_title,
       bk.isbn,
       rs.return_date AS ret_date,
       DATEDIFF(COALESCE(rs.return_date, CURDATE()), iss.issued_date) AS no_of_days
FROM members AS mem
JOIN issued_status AS iss ON mem.member_id = iss.issued_member_id
JOIN books AS bk ON iss.issued_book_isbn = bk.isbn
left JOIN return_status AS rs ON iss.issued_id = rs.issued_id
WHERE rs.return_date is null and DATEDIFF(COALESCE(rs.return_date, CURDATE()), iss.issued_date) >50 order by 1
""", show_res=True)
```

```python
exec__show("""
SELECT 
    ist.issued_member_id,
    m.member_name,
    bk.book_title,
    ist.issued_date,
    -- rs.return_date,
    datediff(CURRENT_DATE, ist.issued_date) as over_dues_days
FROM issued_status as ist
JOIN 
members as m
    ON m.member_id = ist.issued_member_id
JOIN 
books as bk
ON bk.isbn = ist.issued_book_isbn
LEFT JOIN 
return_status as rs
ON rs.issued_id = ist.issued_id
WHERE 
    rs.return_date IS NULL
    AND
    datediff(CURRENT_DATE, ist.issued_date) > 50
ORDER BY 1""", show_res=True)
```
